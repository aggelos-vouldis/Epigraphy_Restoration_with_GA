```python
from typing import List, Tuple

import numpy as np
import os

import matplotlib.pyplot as plt

directory = "./exports"


def get_file_names():
    return os.listdir(directory)


def get_wanted_values(file: str) -> Tuple[List[float], List[int]]:
    best_fittness_values = list()
    total_generations = list()
    f = open(f"{directory}/{file}", 'r', encoding='utf-8')

    while True:
        line = f.readline()
        if not line:
            break
        if "Average Fitness" in line:
            best_fittness_values.append(float(line.split(" --- ")[1].split(" ")[-1][:-1]))
            continue
        if "Total Generations" in line:
            total_generations.append(int(line.split(" ")[2][:-1]))
            continue

    return best_fittness_values, total_generations


def generate_plot(x_values, y_values, title) -> None:
    plt.plot(x_values, y_values)
    plt.title(title)
    plt.show()


if __name__ == "__main__":
    file_names = get_file_names()
    bfv, tg = get_wanted_values(file_names[0])

    for file_name in file_names:
        bfv, tg = get_wanted_values(file_name)
        best_fit_overall = sorted(bfv, reverse=True)[0]
        print(f"{file_name.replace('.log', '')} : {np.mean(bfv):.8f} : {np.mean(tg)} : {best_fit_overall}")
```
